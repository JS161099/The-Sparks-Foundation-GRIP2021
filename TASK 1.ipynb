```python
#Importing the necessasry Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
!conda install -c anaconda seaborn -y
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Reading and Displaying the Data

data = pd.read_csv("http://bit.ly/w-data")
print(data.shape)

data.head()
```

```python
#Visualizing/Plotting the Data

sns.scatterplot(x='Hours', y='Scores', data=data)  
plt.title('Hours vs Percentage')  
plt.xlabel('No. of Hours Studied')  
plt.ylabel('Percentage Scored') 
plt.grid()
plt.show() 
```

```python
#Performing Linear Regression
#Preparing and training the datasets

X = data.iloc[:, :-1].values 
y = data.iloc[:, 1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Training the Algorith of the model

lr = LinearRegression()
lr.fit(X_train, y_train)
coeff = lr.coef_
print(f'The coefficient is {coeff}')
intercept = lr.intercept_
print(f'The intercept is {intercept}')
```

```python
#Plotting the Regression model(line)

line = coeff*X+intercept
plt.scatter(X, y)
plt.plot(X, line, color='r')
plt.show()
```

```python
#Making Predictions

print(X_test) 
y_pred = lr.predict(X_test)
```

```python
#Pictorally comparing Actual and Predicted values

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```
